```python
import keras_tuner as kt

# Specify the directory where the tuner results are saved
tuner_dir = '/content/drive/MyDrive/kt_tuning'

# Re-load the tuner from the saved directory
tuner = kt.Hyperband(
    hypermodel=None,  # No need to specify this, we're just interested in loading past results
    objective='val_accuracy',  # Specify the objective metric to look for (e.g., val_accuracy)
    max_epochs=50,  # Max number of epochs used during the tuning
    directory=tuner_dir,
    project_name='ds2ma_17db_binary'
)

# Get the best hyperparameters based on validation accuracy
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]

# You can now print or use the best hyperparameters
print("Best Hyperparameters:")
print(best_hps.values)

# To get the best validation accuracy
best_trial = tuner.oracle.get_best_trials(num_trials=1)[0]
print(f"Best trial validation accuracy: {best_trial.metrics.get_last_value('val_accuracy')}")
```
