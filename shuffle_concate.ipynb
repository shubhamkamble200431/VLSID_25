```python
import numpy as np

# Load the data
file1 = np.load('/content/drive/MyDrive/signal_-17dB_gaussian_fixed.npy')  # shape (40000, 36, 1000)
file2 = np.load('/content/drive/MyDrive/signal_-17dB_impulsive_fixed.npy')  # shape (40000, 36, 1000)
labels1 = np.load('/content/drive/MyDrive/train/label_-17dB_gaussian.npy')  # shape (40000,)
labels2 = np.load('/content/drive/MyDrive/train/label_-17dB_impulsive.npy')  # shape (40000,)

# Step 1: Vertically concatenate the feature arrays
data = np.vstack((file1, file2))  # shape (80000, 36, 1000)

# Step 2: Reshape the feature array to add a new dimension at the end (for the channel)
data = data[..., np.newaxis]  # shape (80000, 36, 1000, 1)

# Step 3: Concatenate the labels horizontally and reshape them
labels = np.hstack((labels1, labels2))  # shape (80000,)
labels = labels.reshape(-1, 1)  # shape (80000, 1)

# Step 4: Shuffle the data and labels together
indices = np.random.permutation(data.shape[0])
shuffled_data = data[indices]
shuffled_labels = labels[indices]

# Step 5: Save the shuffled data and labels
np.save('/content/drive/MyDrive/17_data.npy', shuffled_data)  # shape (80000, 36, 1000, 1)
np.save('/content/drive/MyDrive/17_labels.npy', shuffled_labels)  # shape (80000, 1)
print(shuffled_data.shape)
print(shuffled_labels.shape)
```
