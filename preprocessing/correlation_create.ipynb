```python
import numpy as np

# Load the .npy file
data = np.load('/content/drive/MyDrive/signal_-17dB_impulsive.npy')  # Shape (40000, 8, 1000)

# Function to calculate cross-correlation and autocorrelation for each data point
def calculate_correlations(data_point):
    receivers = data_point.shape[0]  # Number of receivers (8)
    samples = data_point.shape[1]    # Number of sample points (1000)

    correlations = []

    # Autocorrelation for each receiver
    for i in range(receivers):
        auto_corr = np.correlate(data_point[i], data_point[i], mode='full')[samples-1:]
        correlations.append(auto_corr)

    # Cross-correlation between each pair of receivers
    for i in range(receivers):
        for j in range(i + 1, receivers):
            cross_corr = np.correlate(data_point[i], data_point[j], mode='full')[samples-1:]
            correlations.append(cross_corr)

    return np.array(correlations)

# Initialize the result array
result = np.zeros((data.shape[0], 36, 1000), dtype=complex)

# Loop through each data point and compute the correlation matrix
for idx in range(data.shape[0]):
    result[idx] = calculate_correlations(data[idx])

# Save the result as an .npy file if needed
np.save('/content/drive/MyDrive/signal_-17dB_impulsive_fixed.npy', result)

print("Cross-correlation and autocorrelation calculations complete.")
```
